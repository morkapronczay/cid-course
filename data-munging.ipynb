## FIX COLUMN TYPES

```python
# some basic information
df.info()
```

```python
# try to create a floating point number col
df['Hosszú lejáratú kötelezettségek (eFt)'].astype(float)
```

```python
# check the bad values
for i in df['Adózott eredmény (eFt)']:
    try:
        float(i)
    except:
        print(i)
```

```python
# helper function
def convert_to_float(float_like):
    try:
        return float(float_like)
    except ValueError:
        return float('.'.join(float_like.split(',')[:2]))
```

```python
# gather columns with bad format in a list
bad_columns = ['Üzleti/üzemi eredmény (eFt)'
               , 'Adózott eredmény (eFt)'
               , 'Pénzeszközök (eFt)'
               , 'Saját tőke (eFt)'
               , 'Hosszú lejáratú kötelezettségek (eFt)']
```

```python
# fix them in an iteration
for col in bad_columns:
    df[col] = [convert_to_float(elem) for elem in df[col]]
```

# METRICS

#### HA [teáor = x] AKKOR nem jo MERT nem tud bevételt generálni

```python
# check pl 10 - élelmiszergyártás
df['TEÁOR 08 szerinti teáorszám(ok)'].unique()
```

```python
def check_teaor_big_cat(cat, strict=False, df=df, teaor_col='TEÁOR 08 szerinti teáorszám(ok)'):
    if strict:
        return pd.Series(
            {ind: all([elem.startswith(cat) 
                       for elem in df.loc[ind, teaor_col].split('|')]) 
             for ind in df.index}) 
    else:
        return pd.Series(
            {ind: any([elem.startswith(cat) 
                       for elem in df.loc[ind, teaor_col].split('|')]) 
             for ind in df.index}) 
```

```python
df[check_teaor_big_cat(cat='41', strict=False)].head()
```

```python
df[check_teaor_big_cat(cat='41', strict=True)].head()
```

##### SAVE IT

```python
metrics_df = pd.DataFrame(index=df.index)
```

```python
metrics_df['metrika_1'] = np.where(check_teaor_big_cat(cat='41', strict=True), 1, 0)
```
